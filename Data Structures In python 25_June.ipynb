```
# Sort a given list
mylist=[10,20,5,6,2,8,4,76,87,65,43]

# Sort in Ascending order
mylist.sort()    # Binary sort algorithm :Best case 0(1), Worst case 0(log n)
mylist

# Sort in Descending oder
mylist.sort(reverse=True)
mylist
```

```
# make a matrix for movie recomendation
movie_reviews=[
    [2.5,3.5,3  ,3.5,2.5,3  ],
    [3  ,3.5,1.5,5  ,3.5,3  ],
    [2.5,3  ,0  ,3.5,0  ,4  ],
    [0  ,3.5,3  ,4  ,2.5,4.5],
    [3  ,4  ,2  ,3  ,2  ,3  ],
    [3  ,4  ,0  ,5  ,3.5,3  ],
    [4.5,0  ,0  ,4  ,1  ,0  ]
]

```

```
# Nested list Comprehension
[[user_review[i] for user_review in movie_reviews]for i in range(6)]
```
