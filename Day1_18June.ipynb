```
# CHALLENGE PROBLEM

# part 1.)
# Rabits Wolves Challenge
# Rabbit population=10000 Wolf population=1000

rabits = int(10000)  # rabbit population  
wolves= int(1000)    #  wolf population

a = 0.01     # fractional increase in rabbits population without threat from wolves
b = 0.005    # fractional decrease in wolf population without rabbit to eat
c = 0.00001  # likelihood that a wolf will encounter and eat a rabbit
d = 0.01     # fractional increase in wolf population attributed to a devoured rabbit

print("RABITS POPULATION \t\t WOLVES POPULATION \t\t DAYS")
print("---\t\t------------------\t\t------------------")
for days in range(1,1001):
    rabits = (1+a)*rabits - c*rabits*wolves
    wolves = (1-b)*wolves - c*d*rabits*wolves
    if days % 25 == 0:
        print(rabits,  '\t\t', wolves,  '\t\t', days )
```
