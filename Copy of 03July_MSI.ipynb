```
import pandas as pd
import numpy as np

# Load The Dataset

df = pd.read_csv('http://devanshushukla.com/hackveda.in/videos/telecom_churn.csv')
df.head(10) # Top 10 rows of Dataset
```

```
# check the dimensions
df.shape
```

```
# info() --> used to retrieve the complete summary
# of dataset
df.info()
```

```
# astype() --> used to change the datatype
df['churn'] = df['churn'].astype('int64')
df.info()
```

```
df.head(10)
```

```
# describe() --> used to generate statistical
# summary of Dataset
# generate summary only for numerical features
df.describe()
```

```
# generate statistical summary for non-numerical
# features
df.describe(include=['object','bool'])
```

```
# Sorting operations
# sort_values() --> used to perform sorting

df.sort_values(by='customer service calls').head(10)
```

```
# Retrieve only those records where churn = 1
df[df['churn'] == 1].head(10)

```

```
# How much time (on average) do churn users spend on 
# phone during daytime?

df[df['churn']==1]['total day minutes'].mean()
```

```
df.columns
```

```
# What is the maximum length of international calls amoung loyal
# users(churn==0) who do not have an international plan?

df[df['churn']==0]['total intl minutes'].max()
```

```
df[df['international plan']=='no']['total intl minutes'].max()
```

```
# COMBINING THE ABOVE TWO STATEMENTS:

df[(df['churn']==0)&(df['international plan']=='no')]['total intl minutes'].max()
```

```
# lambda --> anonymous fuctions
# a lambda can take upto n number of arguments
#  SYNTAX:
# lambda argument: expression/condition

x=lambda a:a+5
print(x(60))
```

```
def cube(x):
  return x*x*x

cube(3)
```

```
x=lambda a:a**a
print(x(3))
```

```
# retrieve only those whose initial letter starts with W

df[df['state'].apply(lambda state: state[0]=='W')].head()
```

```
df.head(3)
```

```
# map
# used for mapping 

d={'no':False,'yes':True}
df['international plan']=df['international plan'].map(d)
df.head(10)
```

```
# Replace

df=df.replace({'voice mail plan':d})
df.head(10)
```

```
# add a new column
df['my total charge']=df['total day charge']+\
df['total eve charge']+df['total night charge']+\
df['total intl charge']
df.head()
```

```
# Deleting a column
# axis=1 <-- column
# axis=0 <-- rows
# if you want to update original data make a use of inplace=True
df.drop(['my total charge'],axis=1,inplace=True)
df.head()
```
