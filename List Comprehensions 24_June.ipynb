```
# cubes of a list

cubes=[]
for x in range(5):
  cubes.append(x**3)
cubes
```

```
# List Comprehensions (proram code in a concise way)
# List of 3 line of code in a single line code is known as List Comprehensions:

[x**3 for x in range(5)]   
```

```
# make a list of cube for range 0-4
#Conventional program
cubes =[]
for i in range(10):     # parent Statement
  cubes.append(i**3)   # child Statement
cubes # printing statement
  
```

```
Mylist=[-4,-2,0,2,4]
[x*2 for x in Mylist]
```

```
# Filter negative values out of MylIst
[x for x in Mylist if x>=0]
```

```
# absolute value in Mylist

[abs(x) for x in Mylist]
```

```
# take cube of all the elements of my list and then take the absloute value

[abs(x**3) for x in Mylist]
```

```
# Make a list of Strings
Mylist1=['Salman','Aamir','Ranveer']
[name.strip() for name in Mylist1]
```

```
# Make a list of tuple with(valvue,quad)
# for a range of value 2,4,6,8,10
[(x,x**4) for x in range(2,11,2)]
```

```
Mylist1=[(2,4,6),(1,2,3),(8,9,10)]
[elem for seq in Mylist1 for elem in seq]
```
