```
# Data Gathering
import pandas as pd
credit_data = pd.read_csv("http://13.232.226.32/creditcard.csv")
print(credit_data.columns)
credit_data
```

```
credit_data[["Time", "Amount", "Class"]]
credit_data["Class"].unique() # 0 = Fraud , 1 = Genuine Transaction
```

```
# Lets plot a graph between Amount and Frauds
import matplotlib.pyplot as plt
plt.scatter(credit_data["Class"], credit_data["Amount"])
plt.xlabel("Genuine or Fraud Transactions")
plt.ylabel("Transcation Amount (USD)")
plt.title("Transcations - Amount vs Fraud")
```

```
# Plot a correlations matrix
credit_data.corr()

# Create a correlation matrix
import statsmodels.api as sm
sm.graphics.plot_corr(credit_data.corr(), xnames=credit_data.columns)
```

```
# Second meeting
# 15th October 2019

# A beautiful corelation matrix using seaborn
import seaborn as sb # create high quality heatmaps
plt.subplots(figsize=(40, 10))
sb.heatmap(credit_data.corr() , annot=True)
```

```
credit_data.plot()
```

```
#normalize the data using Standard Score or min-max Scaling
# Formula for min-max scaling = Xnorm =(X-Xmin)/(xmax-Xmin)

x=credit_data
xmin=x.min() # Get the minimize value for each column in credit card data
xmax=x.max() # Get the maximum value for each column in credit card data
xnorm=(x-xmin)/(xmax-xmin)
xnorm #xnorm

#plot the normalized graph
xnorm.plot(figsize=(300,10))
```

```
credit_data.shape

#Sort the data frame in descending order using the column class 
sorted_credit_data=credit_data.sort_values("Class", ascending=False).reset_index()
sorted_credit_data.head(3)
```

```
del sorted_credit_data["index"]
```
