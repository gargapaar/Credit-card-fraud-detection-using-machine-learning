```
#  09th_July_2019- SCRIPT

import pandas as pd
data=pd.read_csv("https://trello-attachments.s3.amazonaws.com/5cf2142046ceb163a0e4b189/5d0739020214d314bd77baf8/69885d162bc4e006add707dc14efb596/Housing_Modified_prepared.csv")
data.plot()
```

```
# Data transformation using Normalization

# Normalization:  https://en.wikipedia.org/wiki/Normalization_(statistics)

# Implement Standardization
# Formula: (X-Xmean)/Xstd
# X= Independent variable
# Xmean= Mean of X variable
# Xstd= Standard Deviation of X independent variable
print("Columns in data",data.columns)
print("Max Value of lotsize before transformation:",data["lotsize"].max())
print("Min Value of lotsize before transformation:",data["lotsize"].min())
print("Mean of lotsize before transformation:",data["lotsize"].mean())

# Perform transformation using STANDARD SCORE (STANDARDIZATION)
X= data["lotsize"]  # Select Independent variable
Xmean=X.mean()      # Calculate the mean of Independent variable
Xstd=X.std()        # Calculate the Standard Deviation of Independent variable 
Xnorm =(X-Xmean)/Xstd
print("X normalized after transformation\n")
Xnorm
print("Min value of Xnorm",Xnorm.min())
print("Max value of Xnorm",Xnorm.max())
```

```
independent_variables = data.columns
independent_variables= independent_variables.delete(0)
print("independent_variables: ",independent_variables)
```

```
X= data[independent_variables]    # independent_variables
Y= data["price"]                  # dependent_variables
X
Y

# Perform Standardization on Independent Vardiables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
Xnorm= scale.fit_transform(X)
print("Normalized data using Standardization")
Xnorm
```

```
# Feature Scaling to normalize data on a common scale (0-1)
# Feature Scaling is also known as Min-Max Scaling
# Formula For Min-Max Scaling is: Xnorm = (X-Xmin)/(Xmax-Xmin)
print("Min value of lotsize: ",data["lotsize"].min())
print("Max value of lotsize: ",data["lotsize"].max())
data["lotsize"].head(2)
```

```
X= data["lotsize"]    # Select the variable
Xmin= X.min()         # Calculate the min value of data variable
Xmax= X.max()         # Calculate the max value of data variable
Xnorm= (X-Xmin)/(Xmax-Xmin)
print("Normalized values after feature scaling\n")
Xnorm
print("Min value for lotsize after transformation",Xnorm.min())
print("Max value for lotsize after transformation",Xnorm.max())
```

```
X= data[independent_variables]
# Feature scale using Min-Max Scaling
Xmin=X.min()
Xmax=X.max()
Xnorm=(X-Xmin)/(Xmax-Xmin)
print("Feature Scale independent variables after transformation\n")
Xnorm

# make a plot of Normalized X
Xnorm.plot()
```

```

#   10th_JULY_2019

# create a machine learning model for house price prediction

# the Linear Regression Equation


import sklearn.linear_model as lm

# Select the linear regression equation
lr=lm.LinearRegression()  # Y= a+b*X

# Train the equation/model
# Calculate values of coefficient of determination (coeff_)[b] and
# Intercept (intercept_)[a] using actual values of X and Y

import pandas as pd
data=pd.read_csv("https://trello-attachments.s3.amazonaws.com/5cf2142046ceb163a0e4b189/5d0739020214d314bd77baf8/69885d162bc4e006add707dc14efb596/Housing_Modified_prepared.csv")
Y= data["price"]   # Taget Data
independent_variables =data.columns
independent_variables =independent_variables.delete(0)
X= data[independent_variables]
lr.fit(X,Y)
print("Intercept:",lr.intercept_)
print("Coefficient of Determination(b):",lr.coef_)

```

```
# pridiction process(calculating Y on the basis of Y=M*X+C)
# Predict the price of the House using Linear Regression

#Ypred = lr.intercept_+(lr.coef_*X)   
Ypred=lr.predict(X)
data["price_predicted"]=Ypred
data[["price","lotsize","price_predicted"]]
```

```
# Take user input and predict price using Linear Model
user_request = {}
for feature in independent_variables:
  temp= input("Enter "+feature+": ")
  user_request[feature]= temp
user_df= pd.DataFrame(data=user_request,index=[0],columns=independent_variables)
#print("User request Dictionary:",user_request)
#print("User request DataFrame:",user_df)
price=lr.predict(user_df)
print("Price of the house is $",round(price[0]))  
```
