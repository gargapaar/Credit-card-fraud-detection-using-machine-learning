```
# 21st JUNE 2019

# DATA ANALYTICS USING PYTHON PROGRAMMING

# Make a list of the elements 10,20,30,40,50
mylist2=[10,20,30,40,50]
mylist2
```

```
# Apppend the Item in the List
mylist2.append(60)
mylist2                                            # complexity is o(1)
```

```
# extend the list with an Interable

iterable=[1,2,3,4]
mylist2.extend(iterable)
mylist2                                            # complexity is o(1)
```

```
# Insert the value at index 6 and vertex 60

mylist2.insert(6,60)
mylist2
```

```
# Remove the Element from the list

mylist2.remove(60)             #complexity is o(n)
mylist2
```

```
# POP method to remove an element from an arbitrary
# Time Complexity O(n)

mylist2.pop(0)  # O(1)
mylist2.pop()   # O(n)
mylist2.pop(2)  # O(n)
mylist2
```

```
# Get the Index of the value provided from the list
mylist=[10,20,30,40]
mylist.index(40)
```

```
# Get the element in the list on a given vertex value
mylist=[10,20,10,40,10,60,10,70]
mylist.count(10)
```

```
# List as a STACK

mystack=[10,20,30,40,50]
mystack.append(60)
mystack
```

```
# List reverse() as Stack

Mylist=[10,20,30,40,50]
Mylist.reverse()
mystack=Mylist
mystack.append(60)
mystack
```

```
# Using List as Queues
from collections import deque   # --> It is important to import everytime when you use the queue
queue=deque([1,2,3,4,5])
queue

```

```
# Add a new member to the queue
queue.append(6)
queue

#  remove a member from the top of the queue
queue.popleft()
queue
```
