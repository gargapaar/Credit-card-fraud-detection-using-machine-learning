```
# read the data
import pandas as pd
data =pd.read_csv("https://trello-attachments.s3.amazonaws.com/5cf2142046ceb163a0e4b189/5d0739020214d314bd77baf8/69885d162bc4e006add707dc14efb596/Housing_Modified_prepared.csv")

# Select Y (dependent variable) and X(Independent variable)
Y=data["price"]
independent_variables=data.columns
independent_variables = independent_variables.delete(0)
X =data[independent_variables]
```

```
# Train the ordinary Least squares regression
from sklearn.linear_model import LinearRegression
from statsmodels.api import OLS

lr=LinearRegression()
lr.fit(X,Y)
```

```
# Train the model using Ordinary Least Squares

from statsmodels.api import OLS
ols=OLS(Y,X).fit()
```

```
# Predict using the Linear Regression Model\
Ypred=lr.predict(X)
Ypred
```

```
# Calculate r2 score for Linear Regression model using
# sum of Squared Total(SST) and Sum of Squared Residual (SSR)

# Calculate Sum of Squared Total (Sum of(y-ymean)^2)
Ymean=Y.mean()
print("Mean of Actual dependent variable",Ymean)

# Calculate square of y - ymean using numpy library
import numpy as np
squared_total= np.square(Y-Ymean)
print("Squared_total is:",squared_total)
sst=squared_total.sum()
print("Sum of squared _total is:",sst)
```

```
# Calculate of sum of squared residual(SSR)
Square_residual=np.square(Ypred-Ymean)
ssr=Square_residual.sum()
print("Sum of squared_residual is:",ssr)
```

```
# Calculate the value of R2 Score(squared)
r2score =ssr/sst
print("R2Score is:",r2score)
```

```
# Calculate of Root Mean Squared Erros(RMSE)
error =Y-Ypred
print("Errors:",error)
# take square of errors
squared_errors=np.square(error)
print("Squared_errors:",squared_errors)
```

```
# calculate mean Squared errors
sse=squared_errors.sum()
mse=squared_errors / len(Y)
print("Mean Squared Errors:",mse)
print("Number of Elements", len(Y))
rmse=np.sqrt(mse)
print("Root mean Suared Errors:",rmse)
```

```
# Calculate Mean Absolute Errors
# Calculate absolute errors
absolute_errors=abs(Y-Ypred)
sae =absolute_errors.sum()
mae=sae/len(Y)
print("Mean Absolute Errors:",mae)
```

```
# Calculate R2Score, RMSE and MAE using sklearn
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r2score=r2_score(Y,Ypred)
print("R2 Score",r2score)
```
